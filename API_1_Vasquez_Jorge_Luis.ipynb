# Analizar la información sobre pasajeros del Titanic

```python
# Importar las librerias necesarias para el análisis
# Importar Pandas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

# Armo el Data Frame con los datos del data set y muestro las primeras 5 filas

```python
# Armando el data frame  
data = pd.read_csv('titanic.csv', sep=';')
# Muestro las primeras 5 filas
data.head()
```

## Verifico si hay datos faltantes en el data set

```python
# Usando la funcion isna() junto con sum() para contar los valores faltantes en cada columna
data.isna().sum()
```

## Reemplazo en la columna pclass todas las filas que estan en Nan con 2

```python
# Reemplazo las filas vacias en pclass con 2.0
data['pclass'].fillna(2, inplace=True)

# verifico los cambios
data.isna().sum()
```

## Reviso y estadarizo los valores de la variable Sex, deben quedar "Male" y "Female"

```python
# Reemplazar las filas de la columna 'sex' que contengan 'Hombre' por 'male' y 'Mujer' por 'female
# Reemplazo Hombre por male
data['sex']=data['sex'].replace('Hombre', 'male')
# Reemplazo Mujer por female
data['sex']=data['sex'].replace('Mujer', 'female')

# verifico los cambios
data.isna().sum() 
```

## Calcular el promedio de edad para reemplazar a los que estan en NaN

```python
# Reemplazo los datos de edad en NaN por el promedio
mean_age = data['age'].mean()
data['age'].fillna(mean_age, inplace=True)

# Verifico si hay datos faltantes en el data set
data.isna().sum()
```

## Acá se debe agregar una 'S' en la variable embarked cuando tenga NaN

```python
# Agreo 'S' en las variables embarcadas que esten en NaN
data['embarked'].fillna('S', inplace=True)

# Verifico si hay datos faltantes en el data set
data.isna().sum()
```

```python
# Exportar a CSV (más común)
data.to_csv('titanic_preprocesado.csv', index=False)

# Con configuración específica
data.to_csv('titanic_preprocesado.csv', 
          index=False, 
          encoding='utf-8',
          sep=',')

print("✅ Archivo CSV exportado: 'titanic_preprocesado.csv'")
```
