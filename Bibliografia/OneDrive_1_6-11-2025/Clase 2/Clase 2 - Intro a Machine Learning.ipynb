# **Preprocesamiento de datasets**

## **Contexto**
Este conjunto de datos proviene originalmente del Instituto Nacional de Diabetes y Enfermedades Digestivas y Renales. El objetivo es predecir, a partir de mediciones diagnósticas, si un paciente tiene diabetes.

## **Contenido**
Se impusieron varias restricciones a la selección de estos casos de una base de datos más amplia. En particular, todos los pacientes aquí son mujeres de al menos 21 años de ascendencia indígena Pima.

## **Variables**

**Pregnancies:** Número de veces que quedó embarazada

**Glucose:** Concentración de glucosa plasmática a las 2 horas en una prueba de tolerancia a la glucosa oral

**BloodPressure:** presión arterial diastólica (mm Hg)

**SkinThickness**: Grosor del pliegue cutáneo del tríceps (mm)

**Insulin**: insulina sérica de 2 horas (mu U/ml)

**BMI:** Índice de masa corporal (peso en kg/(altura en m)^2)

**DiabetesPedigreeFunction:** Función de pedigrí de la diabetes. Es un valor que cuantifica el riesgo genético de desarrollar diabetes basándose en el historial familiar de la enfermedad.

**Age:** Edad (años)

**Outcome:** Variable de clase (0 o 1)

# Gráficos

```python
# -----------------------------------------------------------------------------
# 1. Creación del Countplot (Gráfico de Conteo)
# -----------------------------------------------------------------------------
# Crear el countplot (Gráfico de barras que cuenta automáticamente la frecuencia de cada categoría)
sns.countplot(
    x='Outcome',   # Variable categórica a contar. Seaborn automáticamente cuenta cuántas veces aparece cada valor ('0' o '1') en esta columna.
    data=data      # El DataFrame o conjunto de datos de donde tomar la columna 'Outcome'.
)

# -----------------------------------------------------------------------------
# 2. Configuración del Gráfico (Usando Matplotlib)
# -----------------------------------------------------------------------------
# Añadir título y etiquetas (Estas funciones son de Matplotlib, usadas comúnmente en Seaborn)
plt.title('Distribución de Diabetes (Outcome)') # Establece el título principal del gráfico.
plt.xlabel('Resultado')                        # Etiqueta para el eje horizontal (eje X), representando la variable 'Outcome'.
plt.ylabel('Frecuencia')                       # Etiqueta para el eje vertical (eje Y), representando el conteo de cada resultado.

# -----------------------------------------------------------------------------
# 3. Mostrar el Gráfico
# -----------------------------------------------------------------------------
# Mostrar el gráfico
plt.show() # Renderiza y muestra la figura final con todos los ajustes aplicados.
```

```python
# Paso 1: Contar los valores de la columna 'Outcome'
# La función .value_counts() de pandas cuenta cuántas veces aparece cada valor único en la columna 'Outcome'.
# El resultado es una Serie de pandas donde los índices son los valores únicos (0 y 1) y los valores son sus recuentos o frecuencias.

outcome_counts = data["Outcome"].value_counts()
data["Outcome"].value_counts()
```

```python
# Paso 2: Crear un gráfico de barras para visualizar la distribución

# plt.bar() crea el gráfico de barras.
# - outcome_counts.index: Proporciona las categorías (generalmente 0 y 1) para el eje X.
# - outcome_counts.values: Proporciona las frecuencias (los recuentos) para el eje Y (la altura de las barras).
plt.bar(outcome_counts.index, outcome_counts.values, color=['#1f77b4', '#ff7f0e'])

# plt.xticks() se utiliza para etiquetar los puntos en el eje X con nombres más descriptivos.
# - outcome_counts.index: Indica las posiciones (0 y 1).
# - ['No Diabetes (0)', 'Diabetes (1)']: Define las etiquetas que se mostrarán en esas posiciones.
plt.xticks(outcome_counts.index, ['No Diabetes (0)', 'Diabetes (1)'])

# Título principal del gráfico.
plt.title('Distribución de Diabetes (Outcome)')

# Etiqueta del eje X (qué representa la variable).
plt.xlabel('Resultado')

# Etiqueta del eje Y (qué representan los valores de las barras).
plt.ylabel('Frecuencia')

# Muestra el gráfico en pantalla.
plt.show()
```

## Creación del histograma

```python
# -----------------------------------------------------------------------------
# 1. Creación del Histograma
# -----------------------------------------------------------------------------
plt.hist(
    data["Age"],          # El conjunto de datos a graficar. 'data["Age"]' selecciona la columna 'Age'.
    bins=10,              # Define el número de 'cajas' o intervalos (bins) en los que se dividirá el rango de la edad.
                          # Acá, el rango de edad se divide en 10 barras de igual ancho.
    edgecolor='black'     # Define el color del borde de cada barra para mejorar la separación visual y la claridad.
)

# -----------------------------------------------------------------------------
# 2. Configuración del Gráfico
# -----------------------------------------------------------------------------
plt.title('Histograma')   # Establece el título principal del gráfico, que aparecerá en la parte superior.
plt.xlabel('Edad')        # Etiqueta para el eje horizontal (eje X), que en un histograma representa los valores de la variable (Edad).
plt.ylabel('Frecuencia')  # Etiqueta para el eje vertical (eje Y), que representa el número de observaciones (frecuencia) que caen en cada 'bin' de edad.

# -----------------------------------------------------------------------------
# 3. Mostrar el Gráfico
# -----------------------------------------------------------------------------
plt.show()                # Muestra la ventana del gráfico con todos los elementos configurados.
```

## Revisión de una imputación de datos con el uso de un dataset

```python
# importar numpy
import numpy as np

datos = pd.DataFrame({"id" : [1, 2, 3, 4, 5, 6, 7, 8, 9],
                     "area" : [1, 2, 3, 1, 2, 3, 1, 2, 3],
                     "age" : [32, 30, np.nan, 23, 27, 44, 67, 23, np.nan],
                     "cantidad" : [102, 121, 343, np.nan, 121, np.nan, 155, 149, 221]})
```

```python
datos
```

## Imputación de valores nulos

```python
# Eliminamos los valores nulos
datos.dropna()
```

## Reemplazo de los valores NaN

### - Obteniendo la media de las edades

```python
# media o promedio
media = datos["age"].mean()
datos["age"] = datos["age"].fillna(media)
```
