## Practica de Machine Learning, sobre data set de Diabetes

```python
# importar las bibliotecas de trabajo
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# armo el dateframe de los datos para su análisis
data = pd.read_csv('diabetes.csv')

# ver las primeras filas del dataframe
data.head()
```

```python
# reviso los tipos de datos de cada columna
data.dtypes
```

```python
# reviso si hay datos faltantes
data.isna()
```

```python
# vamos a usar la suma para explorar ya que no sabemos que pasa con los valores centrales
data.isna().sum()
```

```python
# hacemos un histograma 
# ajustamos el tamaño de la fuente
plt.rcParams["font.size"] = 15

# creo una fugura y ajusto su formato
f = plt.figure(figsize=(8,4))

# crear un subplot o subtrama - al ser una sola figura es 1,1,1
ax = f.add_subplot(1,1,1)

# Grafico los datos usando 'hist'. Paso el objeto 'ax' a Pandas. Agrega un borde negro con grosor de 2
data["Outcome"].hist(ax=ax, edgecolor='black', linewidth=2)

# Establezco los limites en el eje X
ax.set_xlim(0, 1)

# establecer la frecuencia de tick. Tengo 0 y 1 que corresponde a SI y NO respectivamente
ax.set_xticks([0, 1])

# etiqueto xtick, labels
ax.set_xticklabels(['N', 'Y'])

# creo el titulo
ax.set_title('Diabetes Y/N')

# establecer la etiqueta del eje X
ax.set_xlabel('Answer')

# establecer la etiqueta del eje Y
ax.set_ylabel('Count')

# establecer los limites superiores e inferiores
ax.set_ylim(0, 510)

# hace que las cosas sean bonitas, pero se adapta al tamaño de la figura
plt.tight_layout()
```
